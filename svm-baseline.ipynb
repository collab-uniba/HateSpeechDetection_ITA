```python
# CARICA IL DATASET
train_set = pd.read_csv("../input/dati-di-training-e-di-test/haspeede2_dev_taskAB.csv", header=None,sep="\t")
test_set  = pd.read_csv("../input/dati-di-training-e-di-test/haspeede2_reference_taskAB-tweets.csv",header=None,sep="\t")
train_set=train_set.drop([0])

# SPLITTA train e test dalle rispettive label
X_train=train_set[1].tolist()
Y_train=[int(x) for x in train_set[2].tolist()]
X_test=test_set[1].tolist()
Y_test=test_set[2].tolist()
```

```python
#FAI IL PREPROCESSING CHE FAI DI SOLITO ekphrases etc etc
vectorizer =TfidfVectorizer(lowercase=True, ngram_range=(1,2), sublinear_tf=True)
dt_matrix = vectorizer.fit_transform(X_train)
test_matrix = vectorizer.transform(X_test)
```

```python
#costruzione SVM


# Creo svm classificatore
# Prova Kernel linear e rbf

clf = dummy.DummyClassifier(strategy='most_frequent')

#Training
clf.fit(dt_matrix, Y_train)

#Predizione
Y_pred = clf.predict(test_matrix)


#valutazione del modello

print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred))

print("Precision:",metrics.precision_score(Y_test,Y_pred))

print("Recall:",metrics.recall_score(Y_test, Y_pred))


# Puoi anche usare classification report (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html)
```
